```python
N, M, K = map(int, input().split())
arr = list(map(int, input().split()))

arr.sort()
arr.reverse()

answer = 0
cnt = 0
for i in range(M) :
    if(cnt < 3) :
        answer = answer + arr[0]
        cnt = cnt + 1
    
    else :
        answer = answer + arr[1]
        cnt = 0

print(answer)
```

```python
N, M, K = map(int, input().split())
arr = list(map(int, input().split()))

arr.sort()
arr.reverse()

answer = 0
cnt = 0

repeat = int(M / (K + 1))
remainder = M % (K + 1)

answer = repeat * (K * arr[0] + arr[1]) + remainder * arr[0]
             
print(answer)
```
