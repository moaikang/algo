```python
import sys
sys.setrecursionlimit(100000000)

n = int(input())
triangle = []
results = 0

for i in range(n) :
    triangle.append(list(map(int, input().split(' '))))

def dfs(sum, depth, idx) :
    global results
    global triangle
    
    if(depth == n) : 
        if(results < sum) :
            results = sum
            return
        return
    
    dfs(sum + triangle[depth][idx], depth + 1, idx)
    dfs(sum + triangle[depth][idx], depth + 1, idx + 1)
        
dfs(0, 0, 0)    
print(results)
```

### 위의 코드는 그냥 순수 DFS, 왜 메모리 초과 일까?
