```python
import sys
sys.setrecursionlimit(10**6)

T = int(input()) # 테스트의 개수
cabbageMap = []
checked = []

dx, dy = [0, 1, 0, -1], [1, 0, -1, 0]

def dfs(x, y) :
    global checked, dx, dy, cabbageMap
    checked[x][y] = True
    for i in range(4) :
        nextX = x + dx[i]
        nextY = y + dy[i]
        
        if cabbageMap[nextX][nextY] == 0 or checked[nextX][nextY] == True :
            continue
        dfs(nextX, nextY)
                
def process() :
    global checked, dx, dy, cabbageMap
    answer = 0
    M, N, K = map(int, input().split(' '))
    cabbageMap = [[0 for _ in range(M + 2)] for __ in range(N + 2)]
    checked = [[False for _ in range(M + 2)] for __ in range(N + 2)]

    for _ in range(K) :
        x, y = map(int, input().split(' '))
        cabbageMap[y + 1][x + 1] = 1
        
    for i in range(1, N + 1) :
        for j in range(1, M + 1) :
            if(cabbageMap[i][j] == 0 or checked[i][j]): 
                continue
            dfs(i, j)  
            answer = answer + 1
    print(answer)

for _ in range(T) :
    process()
```
