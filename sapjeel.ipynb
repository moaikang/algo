```python
def isBalancedBrackets(brackets) :
    left = 0
    right = 0
    for bracket in brackets :
        if(bracket == '(') :
            left = left + 1
        else :
            right = right + 1
        
    if left == right :
        return True
    return False

def isValidBrackets(brackets) :
    stack = []
    
    for bracket in brackets :
        if bracket == '(' :
            stack.append(bracket)
        else :
            if len(stack) == 0 :
                return False;
            stack.pop()
        
    if(len(stack) == 0) :
        return True
    return False
        
    
def divide(p) : # 2
    if p == '' : # 입력이 빈 문자열인 경우
        return ''
    u = ''
    v = ''
    
    for i in range(len(p)) :
        if p[i] == '(' :
            u = u + '('
        else :
            u = u + ')'
        
        if isBalancedBrackets(u) and isBalancedBrackets(p[i + 1: len(p)]) :
            v = p[i + 1: len(p)]
            
            if isValidBrackets(u) :
                return u + divide(v)
            else :
                temp = '('
                temp = temp + divide(v)
                temp = temp + ')'
                
                u = u[1 : len(u) - 1]
                reversedU = ''
                for bracket in u :
                    if u == '(' :
                        reversedU = reversedU + ')'
                    else :
                        reversedU = reversedU + '('
                    
                return temp + reversedU    
    

def solution(p):
    return divide(p)

isValidBrackets("()))((()")
```

```python
from collections import deque

N, M, K, X = map(int, input().split()) #도시, 간선, 거리, 시작 도시

def bfs(graph, first) :
    global K

    queue = deque()
    queue.append((first, 0))
    answer = []
    
    while(queue) :
        v, move = queue.popleft()
        visited[v] = True
        if move == K :
            answer.append(v)
            
        cities = graph[v]
        
        for city in cities :
            if(visited[city] == True) :
                continue
                
            queue.append((city, move + 1))
            visited[city] = True
            
    return answer
        
graph = [[] for i in range(1, N + 2)]
visited = [False for i in range(0, N + 2)]
visited[0] = True

for i in range(M) :
    A, B = map(int, input().split())
    graph[A].append(B)
    
cities = bfs(graph, X)

if len(cities) == 0 :
    print(-1)

else :
    for city in sorted(cities) :
        print(city)




    
```

```python
N, K = list(map(int, input().split())) # N = 가로, 세로 | K = K 이하의 수만 바이러스로 주어진다.

graph = []

for i in range(N) :
    graph.append(list(map(int, input().split())))
    
S, X, Y = map(int, input().split()) # S = 몇초 뒤에, X, Y 자리에 뭐가 있을지

dx, dy = [0, 0, -1, 1], [1, -1, 0, 0] # 상 하 좌 우

def isAllExaminerPropagated(graph) :
    for i in range(N) :
        for j in range(N) :
            if graph[i][j] == 0 :
                return False
    return True
            

def propagate(graph) :
    global dx, dy
    visited = [[False for j in range(N)] for i in range(N)]
    
    for i in range(N) :
        for j in range(N) :
            
            if visited[i][j] :
                continue
            
            if graph[i][j] == 0 :
                continue
                
            for next in range(4) :
                nx = i + dx[next]
                ny = j + dy[next]
                
                if nx < 0 or nx >= N or ny < 0 or ny >= N : # 인덱스를 벗어났을 때
                    continue
                    
                if graph[nx][ny] == 0 : # 아직 침략 당하지 않았을 때
                    visited[nx][ny] = True
                    graph[nx][ny] = graph[i][j]
                    continue
                
                if visited[nx][ny] : # 침략 당했을 때
                    if graph[i][j] < graph[nx][ny] : # 더 작다면, 그걸로 바꿀꺼야
                        graph[nx][ny] = graph[i][j]
                        
    return graph
     
isDone = False
for i in range(S) :
    if isAllExaminerPropagated(graph) :
        break
        
    graph = propagate(graph)

print(graph[X - 1][Y - 1])
                
```

```python
import copy

dx, dy = [0, 0, -1, 1], [1, -1, 0, 0] # 상 하 좌 우

def canStudentsHide(graph, N) :
    global dx, dy
    
    for i in range(N) :
        for j in range(N) :
            if graph[i][j] == 'T' :
                canHide = True
                for nextVector in range(4) :
                    for nextSize in range(1, N) :
                        nx = i + (dx[nextVector] * nextSize)
                        ny = j + (dy[nextVector] * nextSize)
                        
                        if not isValidIndex(nx, ny, N) :
                            break
                            
                        if graph[nx][ny] == 'S' :
                            return False
                        
                        if graph[nx][ny] == 'O' or graph[nx][ny] == 'T' :
                            break
    return True
                
def getObstacleCandidates(graph, N) :
    global dx, dy
    candidates = []
    
    for i in range(N) :
        for j in range(N) :
            if graph[i][j] == 'T' :
                for nextVector in range(4) :
                    for nextSize in range(1, N) :
                        nx = i + (dx[nextVector] * nextSize)
                        ny = j + (dy[nextVector] * nextSize)
                        
                        if not isValidIndex(nx, ny, N) :
                            break
                        
                        if (nx == 0 or ny == 0 or nx == N - 1 or ny == N - 1) and graph[nx][ny] == 'X' :
                            for n in range(1, nextSize + 1) :
                                defaultNx = i + (dx[nextVector] * n)
                                defaultNy = j + (dy[nextVector] * n)
                                candidates.append((defaultNx, defaultNy))
                            break
                            
                        if graph[nx][ny] == 'S' or graph[nx][ny] == 'T' :
                            for n in range(1, nextSize) :
                                defaultNx = i + (dx[nextVector] * n)
                                defaultNy = j + (dy[nextVector] * n)
                                candidates.append((defaultNx, defaultNy))
                            break
                                                        
    return candidates             
    
def isValidIndex(x, y, N) :
    if x < 0 or x >= N or y < 0 or y >= N :
        return False
    return True
       
def process() :
    global dx, dy
    
    N = int(input()) # 행, 열
    graph = []

    for i in range(N) :
        graph.append(list(input().split()))
        
    # 선생 바로 옆에 있는지를 체크 (장애물 넣을 틈도 없음)
    for i in range(N) :
        for j in range(N) :
            if graph[i][j] == 'T' :
                for next in range(4) :
                    nx = i + dx[next]
                    ny = j + dy[next]
                
                    if not isValidIndex(nx, ny, N) :
                        continue
                    
                    if graph[nx][ny] == 'S' :
                        return 'NO'
    
    candidates = getObstacleCandidates(graph, N)
            
    for i in range(len(candidates)) :
        for j in range(i + 1, len(candidates)) :
            for k in range(j + 1, len(candidates)) :
                copiedGraph = copy.deepcopy(graph)
                x1, y1 = candidates[i][0], candidates[i][1]
                x2, y2 = candidates[j][0], candidates[j][1]
                x3, y3 = candidates[k][0], candidates[k][1]
                
                copiedGraph[x1][y1] = 'O'
                copiedGraph[x2][y2] = 'O'
                copiedGraph[x3][y3] = 'O'
                
                if(canStudentsHide(copiedGraph, N)) :
                    return 'YES'
                
    return 'NO'
                

print(process())
    
    
        
```

```python
def findIndexOfNumber(keyPad, target) :
    for i in range(len(keyPad)) :
        for j in range(len(keyPad[i])) :
            if keyPad[i][j] == target :
                return (i, j)

def calculateDistance(keyPad, target, origin) :
    targetX, targetY = findIndexOfNumber(keyPad, target)
    originX, originY = findIndexOfNumber(keyPad, origin)
    
    return abs(targetX - originX) + abs(targetY - originY)
            
def solution(numbers, hand):
    keyPad = [[1, 2, 3], [4, 5, 6], [7, 8, 9], ['*', 0, '#']]
    mustLeftKeys = [1, 4, 7]
    mustRightKeys = [3, 6, 9]
    
    leftHand = (3, 0)
    rightHand = (3, 2)
    
    answer = ''
    
    for number in numbers :
        if number in mustLeftKeys :
            x, y = findIndexOfNumber(keyPad, number)
            leftHand = (x, y)
            answer = answer + 'L'
            continue
        
        if number in mustRightKeys :
            x, y = findIndexOfNumber(keyPad, number)
            rightHand = (x, y)
            answer = answer + 'R'
            continue
        
        # 상황에 따라서 달라지는 키패드일때
        leftDistance = calculateDistance(keyPad, number, keyPad[leftHand[0]][leftHand[1]])
        rightDistance = calculateDistance(keyPad, number, keyPad[rightHand[0]][rightHand[1]])
        
        x, y = findIndexOfNumber(keyPad, number)
        
        if leftDistance > rightDistance :
            leftHand = (x, y)
            answer = answer + 'L'
            
        if leftDistance == rightDistance :
            if hand == 'left' :
                leftHand = (x, y)
                answer = answer + 'L'
            
            else :
                rightHand = (x, y)
                answer = answer + 'R'
                
        else :
            rightHand = (x, y)
            answer = answer + 'R'
        
    return answer

solution([1, 3, 4, 5, 8, 2, 1, 4, 5, 9, 5], 'right')
```
